### 3. Методи класу

```python
class MyAnimals:
    def __init__(self, animals, name, age, mass = None) -> None:
        print(f"В мене є тваринка {animals} яку звати {name} які {age} років")
        self.animals = animals
        self.name = name
        self.age = age
        self.mass = 0 if mass==None else mass
    
    def speak(self):
        if self.animals == "Кіт":
            return f"Моя Тваринка {self.animals} говорить Мяууу"
        elif self.animals == "Собака":
            return f"Моя Тваринка {self.animals} говорить Гавв"
        else:
            return f"Моя Тваринка {self.animals} не розмовляє"
    
    def add_one_year(self):
        self.age += 1
        return f"Моя Тваринка {self.animals} прожила за мною один рік"

a1 = MyAnimals("Рибки", "Неони", 1)
a2 = MyAnimals("Кіт", "Мурка", 5, 10)
a3 = MyAnimals(animals="Собака", name="Рекс", age=2)

print(f"Тваринка {a1.animals}:", a1.speak())

print(f"Тваринка {a3.animals}:", a3.speak())


print(f"{a1.animals} живе зі мною {a1.age} років")
print(a1.add_one_year())
print(f"{a1.animals} живе зі мною {a1.age} років")

print(a1.age)
```

- Завдання задати декілька обєктів в циклі та вивести їх глобальну змінну

```python
class MyAnimals:
    # змінні класу (глобальні для всіх об’єктів)
    total_animals = 0
    total_animal_types = set()

    def __init__(self, animals, name, age, mass=None) -> None:
        print(f"В мене є тваринка {animals} яку звати {name} які {age} років")
        self.animals = animals
        self.name = name
        self.age = age
        self.mass = 0 if mass is None else mass
        # оновлюємо глобальні змінні
        MyAnimals.total_animal_types.add(animals)
        MyAnimals.total_animals += 1


# список даних для створення об’єктів
animal_data = [
    ("Рибки", "Неони", 1),
    ("Рибки", "Гупії", 1),
    ("Рибки", "Соміки", 1),
    ("Кіт", "Мурка", 5, 10),
    ("Собака", "Рекс", 2)
]

# створюємо об’єкти у циклі
animals = []
for data in animal_data:
    animals.append(MyAnimals(*data))

# виводимо глобальні змінні класу
print(f"В мене зараз є тваринки {list(MyAnimals.total_animal_types)}, їх кількість {MyAnimals.total_animals}")
```

### 5. Статичні методи та змінна self

```python
class MyAnimals:
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age
    
    def add_one_year(self):
        self.age += 1
        return f"Моя Тваринка {self.animals} прожила за мною один рік"

    @staticmethod
    def call_pet():
        return "Тваринка біжить до нас"


a1 = MyAnimals("Рибки", "Гупії", 1)
a2 = MyAnimals("Кіт", "Мурка", 5)

print(f"Ми кличемо нашу тваринку за допомогою статичного методу з обєкту |{a1.call_pet()}| та з класу |{MyAnimals.call_pet()}|")

print(f"Викликаю метод з обєкту |{a1.add_one_year()}| та з класу |{MyAnimals.add_one_year(a1)}|")
```

```python
from datetime import datetime
class MyAnimals:
    max_age = 20
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age
    
    @property
    def live_untill(self):
        return datetime.today().year - self.age + self.max_age
    
    def add_one_year(self):
        self.age += 1
        return f"Моя Тваринка {self.animals} прожила за мною один рік"


a1 = MyAnimals("Кіт", "Мурка", 5)
print(f"Мій {a1.animals} будже зі мною до {a1.live_untill} року")
a1.add_one_year()
print(f"Мій {a1.animals} будже зі мною до {a1.live_untill} року")
```

- Завдання додати ще одну properties та вивести її

```python
from datetime import datetime

class MyAnimals:
    max_age = 20
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age
    
    @property
    def live_untill(self):
        return datetime.today().year - self.age + self.max_age
    
    @property
    def is_adult(self):
        return self.age >= 2
    
    def add_one_year(self):
        self.age += 1
        return f"Моя Тваринка {self.animals} прожила за мною один рік"


# створюємо об’єкт
a1 = MyAnimals("Кіт", "Мурка", 5)

# використовуємо properties
print(f"Мій {a1.animals} буде зі мною до {a1.live_untill} року")
print(f"Чи моя тваринка доросла? {a1.is_adult}")

# додаємо рік і перевіряємо ще раз
a1.add_one_year()
print(f"Мій {a1.animals} буде зі мною до {a1.live_untill} року")
print(f"Чи моя тваринка доросла? {a1.is_adult}")
```

### 7. Методи Класу

```python
class MyAnimals:
    total_csv = int(0)
    total_objects = int(0)
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age
        MyAnimals.total_objects += 1

    @classmethod
    def from_csv(cls, csv_data:str):
        cls.total_csv += 1
        animal, name, age = csv_data.split(",")
        return cls(animal, name, age)

a1 = MyAnimals.from_csv("Кіт,Мурка,5")
a2 = MyAnimals("Собака", "Рекс", 2)

print(f"Мій {a1.animals}, звати {a1.name}, має {a1.age} років")
print(f"Мій {a2.animals}, звати {a2.name}, має {a2.age} років")
print(f"Загальна кількість обєктів {MyAnimals.total_objects} а створених з csv формату {MyAnimals.total_csv}")
```

- Завдання дописати метод класу який буде превіряти що атрибут name заданий з великої букви 

```python
class MyAnimals:
    total_csv = int(0)
    total_objects = int(0)

    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = int(age)   # щоб age був числом
        MyAnimals.total_objects += 1

    @classmethod
    def from_csv(cls, csv_data: str):
        cls.total_csv += 1
        animal, name, age = csv_data.split(",")
        return cls(animal, name, age)

    @classmethod
    def check_name_capitalized(cls, obj):
        """Перевіряє, чи атрибут name починається з великої букви"""
        return obj.name[0].isupper()

a1 = MyAnimals.from_csv("Кіт,Мурка,5")
a2 = MyAnimals("Собака", "Рекс", 2)
a3 = MyAnimals("Рибки", "неони", 1)  # спеціально з малої букви

print(f"Мій {a1.animals}, звати {a1.name}, має {a1.age} років")
print(f"Мій {a2.animals}, звати {a2.name}, має {a2.age} років")
print(f"Мій {a3.animals}, звати {a3.name}, має {a3.age} років")

print(f"Загальна кількість об’єктів {MyAnimals.total_objects}, а створених з csv формату {MyAnimals.total_csv}")

print(f"Ім’я {a1.name} з великої букви? {MyAnimals.check_name_capitalized(a1)}")
print(f"Ім’я {a2.name} з великої букви? {MyAnimals.check_name_capitalized(a2)}")
print(f"Ім’я {a3.name} з великої букви? {MyAnimals.check_name_capitalized(a3)}")
```

### Магічні методи

```python
class MyAnimals:
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age

    def __repr__(self) -> str:
        return f'MyAnimals("{self.animals}", "{self.name}", {self.age})'

    def __str__(self) -> str:
        return f"Це є обєкт {self.__repr__()}"

a1 = MyAnimals("Собака", "Рекс", 2)
print(a1)

print(repr(a1))

str(a1)
```

### Арифметичні методи

```python
a = 2
b = 5
c = "C"

print(a+b, a - int(1),  a * b, str(a) + c)

print(a.__add__(b), a.__sub__(int(1)), a.__mul__(b))

class MyAnimals:
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age

    def __sub__(self, obj):
        if isinstance(obj, MyAnimals):
            return f"Мій {self.name} покусав {obj.name}"
        elif isinstance(obj, int):
            self.age += obj
            return f"Мій {self.name} постарів на {obj} років, тепер йому {self.age}"
        return NotImplementedError
    
    def __gt__(self, obj):
        if isinstance(obj, MyAnimals):
            if self.age > obj.age:
                return f"Старшим є {self.name}"
            return f"{self.name} не є старшим за {obj.name}"
        return NotImplementedError

    def __len__(self) -> int:
        print("Ми зайшли в метон len, та знаходимо довжину Імені тваринки")
        return len(self.name)
            
        

a = MyAnimals("Собака", "Рекс", 2)
b = MyAnimals("Кіт", "Мурзик", 4)
print(a - b)
print(a - 1)
print(a - float(1))

print(a > b)

print(b > a)

len(a)
```
